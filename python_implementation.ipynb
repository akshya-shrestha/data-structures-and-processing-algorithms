```python
# initialising matrices B and C

B = np.array([[4,7,2],[3,2,5],[6,4,3]])
C = np.array([[3,1,9],[7,5,8],[2,1,1]])

# calculating matrix product using
# python function

# multiplication result is 3x3 matrix
result = [[0,0,0],
         [0,0,0],
         [0,0,0]]

# iterating through rows of B
for i in range(len(B)):
   # iterating through columns of C
   for j in range(len(C[0])):
       # iterating through rows of C
       for k in range(len(C)):
            result[i][j] += B[i][k] * C[k][j]
            
print("The multiplication is: ")        
for r in result:
    print(r)
```

```python
# product of matrices B and C using numpy libraray
P = np.matmul(B,C)
print("The product of matices B and C is: ")
print(P)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Store the dataset into a data frame
ds = pd.read_csv('task3_average_monthly_temperatures.csv', delimiter = ',')
df = pd.DataFrame(ds)

def DrawBox(x, y, size, r, g, b):

    if r < 0:

        r = int(0)

    if g < 0:

        g = int(0)

    if b < 0:

        b = int(0)

    if r > 255:

        r = int(255)

    if g > 255:

        g = int(255)

    if b > 255:

        b = int(255)

    for i in range(0, int(size)):

        plt.plot([x, x + size], [y + i, y + i], '#{:02x}{:02x}{:02x}'.format(r, g, b))



# Set the plot

plt.figure(figsize=(18, 3.5))

# plt.axis([0, 600, 0, 400])

plt.xticks([])

plt.yticks([])

plt.axis('off')



Min = int(min(df.min(numeric_only=True)))

Max = int(max(df.max(numeric_only=True)))

BoxSize = int(40)

OffsetX = int(15)

OffsetY = int(12)

# Allowing user to give threshold value
#Input threshold

Threshold = int(input("Input Threshold: "))

# Generate the heat map

for i in range(0, df.shape[0]):

    for j in range(1, df.shape[1]):

        ColourCode = int(((df.values[i, j]-Min)/(Max-Min))*255)

        if df.values[i, j] > Threshold:

          DrawBox(20+BoxSize*j, 300-BoxSize*i, BoxSize, 0, 0, ColourCode)

        if df.values[i, j] <= Threshold:

          DrawBox(20+BoxSize*j, 300-BoxSize*i, BoxSize, ColourCode, 0, 0)

        plt.text(OffsetX+20+BoxSize*j, OffsetY+300-BoxSize*i, str(df.values[i, j]), color='white')




for i in range(0, 256):

    plt.plot([560, 580], [i + 85, i + 85], '#{:02x}{:02x}{:02x}'.format(int(i), 0, 0))
    plt.text(585, 80, Min)
    plt.text(585, 320, Max)
    plt.text(72, 160, 'Jan', fontsize=13)
    plt.text(112, 160, 'Feb', fontsize=13)
    plt.text(152, 160, 'Mar', fontsize=13)
    plt.text(192, 160, 'Apr', fontsize=13)
    plt.text(232, 160, 'May', fontsize=13)
    plt.text(272, 160, 'Jun', fontsize=13)
    plt.text(312, 160, 'Jul', fontsize=13)
    plt.text(352, 160, 'Aug', fontsize=13)
    plt.text(392, 160, 'Sep', fontsize=13)
    plt.text(432, 160, 'Oct', fontsize=13)
    plt.text(472, 160, 'Nov', fontsize=13)
    plt.text(512, 160, 'Dec', fontsize=13)
    plt.text(5, 315, str(df.City[0]), fontsize=13)
    plt.text(5, 275, str(df.City[1]), fontsize=13)
    plt.text(5, 235, str(df.City[2]), fontsize=13)
    plt.title('Heatmap for Monthly Average Temperature of Cities')


plt.show()
```
